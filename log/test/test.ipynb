```python
fs = ! ls

fs = [f for f in fs if 'ckpt' in f]
print(fs)
```

```python
import json
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
for f in fs:
    fi = ! ls $f
    path = f+'/'+fi[0]
    print(path)
    fii = ! ls $path
    path = path +  '/'+fii[1]
    print(path)
    with open(path) as f:
        d = json.load(f)
        print(d.keys())
    # plt.plot(d['episode_reward']['env0_first_0'], label=fi)
    plt.plot(d['episode_length'], label=fi)
plt.legend()
    
```
