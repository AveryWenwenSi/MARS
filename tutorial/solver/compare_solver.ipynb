```python
from matplotlib import pyplot as plt
import numpy as np

oracle_v_star = -0.296071
solvers = ['DoubleOralce', 'OracleDoubleOralce', 'FictitiousSelfPlay', 'OracleFictitiousSelfPlay', 'NashVI', 'NashVI_MWU', 'NashQLearning']
legends = ['Double Oracle', '(Oracle) Double Oracle', 'Fictitious Selfplay', '(Oracle) Fictitious Selfplay', 'Nash VI (ECOS)', 'Nash VI (MWU)', 'Nash Q-Learning']
plt.figure(figsize=(8, 6), dpi=300)
xlabel_multiply = 10 # get one exploitability value per 10 episodes
for s, l in zip(solvers, legends):
    path = f'./data/{s}/data.npy'
    exp_data = np.load(path)
    x=np.arange(len(exp_data))*xlabel_multiply
    plt.plot(x, exp_data+oracle_v_star, label=l)

plt.title('Comparison')
plt.xlabel('episodes')
plt.ylabel('exploitability')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid()
plt.savefig('./data/compare.png')
plt.show()


```
